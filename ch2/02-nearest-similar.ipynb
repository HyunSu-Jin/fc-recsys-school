```python
landmark_ds = pd.DataFrame.from_csv('/Users/goodvc/data/fc-recsys-school/ch2/resource/landmark.csv')
landmark_ds.head()
```

```python
from scipy.spatial import distance

## 유사도를 측정
def most_similar(name, ds, distance_func=distance.euclidean):
    v1 = ds.loc[name]
    ret = []
    for name, row in ds.iterrows():
        ret.append( [name, round(distance_func(v1,row),5)] )
    return pd.DataFrame(ret,columns = ['name','distance']).sort(['distance'], ascending=True)
```

최규민1과 비슷한 사람은?

```python
most_similar('wonbin:1', landmark_ds, distance.euclidean)
```

유사도 함수별 측정해 보자 
* euclidean, cosine, correlation

```python
target='goodvc:01'

for func in [distance.euclidean, distance.cosine, distance.correlation]:
    print( func.__name__, " ***************\n"
          ,most_similar(target,landmark_ds, func))        
```

## 유저 유사도 메트릭스

```python
users = landmark_ds.index
func = distance.euclidean

sim_matrix = []
for u1 in users:
    similarites = []
    u1_vec = landmark_ds.loc[u1].values.tolist()
    for u2 in users:
        u2_vec = landmark_ds.loc[u2].values.tolist()
        similarites.append(round(func(u1_vec, u2_vec),3))
    sim_matrix.append(similarites)

sim_matrix = pd.DataFrame(sim_matrix, index=users, columns=users)
sim_matrix
```

```python
def make_sim_matrix(distance_func=distance.euclidean):
    users = landmark_ds.index

    sim_matrix = []
    for u1 in users:
        similarites = []
        u1_vec = landmark_ds.loc[u1].values.tolist()
        for u2 in users:
            u2_vec = landmark_ds.loc[u2].values.tolist()
            similarites.append(round(distance_func(u1_vec, u2_vec),5))
        sim_matrix.append(similarites)

    sim_matrix = pd.DataFrame(sim_matrix, index=users, columns=users)
    return sim_matrix
```

### 가장 유사한 사용자 찾기

```python
distance_func=distance.euclidean
sim_matrix=make_sim_matrix(distance_func)
print( distance_func.__name__ , "Distance Function")
for idx,row in sim_matrix.iterrows():
    row.sort()
    print(idx, row[1:3].index.values)
```

```python
for distance_func in [distance.euclidean, distance.cosine, distance.correlation]:
    sim_matrix=make_sim_matrix(distance_func)
    print( distance_func.__name__ , "Distance Function*******")
    for idx,row in sim_matrix.iterrows():
        row.sort()
        print(idx, row[1:3].index.values)
 
```
